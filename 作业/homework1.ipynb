1、打开任意一幅灰度图片，计算该图像的图像直方图：

首先导入相关库，并读取图片。

```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import math
 
img = Image.open('image.bmp')#选择自己的图片路径导入
s=np.array(img.convert('L'))
 
plt.imshow(s, cmap='gray')#彩色图转灰度图
```

```python
def Hist(img):
    #补充相关功能，计算图像尺寸，像素个数，灰度级个数，绘制灰度直方图
    height, width = s.shape
    histogram = np.array([0] * 256)
    for i in range(height):
        for j in range(width):
            index = s[i][j]
            histogram[index] += 1
    plt.bar(np.arange(256), histogram)
    plt.xlabel("gray label")
    plt.ylabel("number of pixels")
    plt.axis([0, 255, 0, np.max(histogram)])
    plt.show()
    
```

```python
Hist(s)
```

2、打开任意一幅灰度图片，对该图像进行处理，使得整幅图像变亮10个灰度级：

```python
def grayboost(img):
    #补充相关功能
    s = np.array(img.convert('L'))
    s = s.astype(np.uint16)
    s += 10
    print(np.max(s), np.min(s))
    s_clipped = np.clip(s, 0, 255).astype(np.uint8)
    plt.imshow(s_clipped, cmap='gray')
```

```python
grayboost(img)
```

3、打开任意一幅灰度图片，在该图像的中心位置画一个半径为半个图像高度或宽度的圆：

```python
def drawcircle(img):
    #补充相关功能
    s = np.array(img.convert('L'))
    r = min(s.shape) / 2
    thickness = 1
    h, w = s.shape
    ox = int(w/2)
    oy = int(h/2)
    for i in range(h):
        for j in range(w):
            d = math.hypot(i - oy, j - ox)
            if d <= r and d >= r-thickness:
                s[i][j] = 255
    plt.imshow(s, cmap='gray')
```

```python
drawcircle(img)
```
